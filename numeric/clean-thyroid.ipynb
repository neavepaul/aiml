```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('thyroidDF.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
# Remove duplicates if any
df.drop_duplicates(inplace=True)
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print(missing_values)
```

```python
df.replace('?',np.nan,inplace=True)
```

```python
df
```

```python
df.median()
```

```python
df.sex.mode()
```

```python
df.sex.fillna(('F'),inplace=True)
df.age.fillna(int('54'),inplace=True)
df.TSH.fillna(float('1.40'),inplace=True)
df.T3.fillna(float('2'),inplace=True)
df.TT4.fillna(float('103'),inplace=True)
df.T4U.fillna(float('0.98'),inplace=True)
df.FTI.fillna(float('104'),inplace=True)
df.drop(["TBG"],axis=1,inplace=True)
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print(missing_values)
```

```python
df['age']=df['age'].astype("int")
df['FTI']=df['FTI'].astype("float")
df['T4U']=df['T4U'].astype("float")
df['TT4']=df['TT4'].astype("float")
df['TSH']=df['TSH'].astype("float")
df['T3']=df['T3'].astype("float")
```

```python
Q1= df.age.quantile(0.25)
Q2= df.age .quantile(0.50)
Q3= df.age.quantile(0.75)
IQR = Q3 - Q1
lowerend = Q1 - (1.5 * IQR)
upperend = Q3 + (1.5 * IQR)

dfdia_outliers =df.age[(df.age < lowerend) | (df.age>upperend)]

df.age= np.where(df.age>upperend ,upperend,df.age)
df.age= np.where(df.age <lowerend ,lowerend,df.age )
```
